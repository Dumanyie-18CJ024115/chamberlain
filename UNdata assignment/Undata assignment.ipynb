```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics
import math


africanCountries=["Algeria", "Angola", "Benin", "Botswana", "Burkina", "Burundi", 
                  "Cameroon", "Cape Verde", "Central African Republic", "Chad","Comoros",
                  "Congo", "Congo, Democratic Republic of", "Djibouti", "Egypt", 
                  "Equatorial Guinea", "Eritrea", "Ethiopia", "Gabon", "Gambia", 
                  "Ghana", "Guinea", "Guinea-Bissau", "Ivory Coast", "Kenya", 
                  "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", 
                  "Mali", "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", 
                  "Niger", "Nigeria", "Rwanda", "Sao Tome and Principe", 
                  "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", 
                  "South Sudan", "Sudan", "Swaziland", "Tanzania", "Togo", "Tunisia", "Uganda", "Zambia", "Zimbabwe"]

data= pd.read_csv('Book1.csv')
data=data.set_index('Entity')
data
```

```python
for i in range(3):
    l = 2010+i*3
    h = l + 3
    fig, axes  = plt.subplots(1,3, figsize=(10,30), sharey=True,)
 
    for idx, year in enumerate(range(l,h)):
        data[data['Year'] == year]['Value'].plot.barh(ax=axes[idx])
        axes[idx].set_title(year)
```

```python
theMean = []
for year in range(2010,2019):
    theMean.append(data[data['Year'] == year]['Value'].mean())
    
plt.plot(range(2010,2019), theMean)
plt.title('The Mean of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('Mean Plot')
```

```python
theMedian = []
for year in range(2010,2019):
    theMedian.append(data[data['Year'] == year]['Value'].median())
    
plt.plot(range(2010,2019), theMedian)
plt.title('The Median of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('Median plot')
```

```python
variance = []
for year in range(2010,2019):
    variance.append(data[data['Year'] == year]['Value'].var())
    
plt.plot(range(2010,2019), variance)
plt.title('The Variance of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('Variance plot')
```

```python
sD = []
for year in range(2010,2019):
    sD.append(data[data['Year'] == year]['Value'].var()**0.5)
    
plt.plot(range(2010,2019), sD)
plt.title('The standard deviation of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('Standard Deviation plot')
```

```python
interq = []
def get_iqr(sample):
    q3,q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

for year in range(2010,2019):
    interq.append(get_iqr(data[data['Year'] == year]['Value']))
    
    
plt.title('The interquartile range of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('interquartile plot')
plt.plot(range(2010,2019), interq)
```

```python
s_interq = []

for year in range(2010,2019):
    s_interq.append(get_iqr(data[data['Year'] == year]['Value'])*0.5)
    
    
plt.title('The semi interquartile range of Female to Male birth ratio from 2010 to 2019')
plt.xlabel('Year')
plt.ylabel('semi interquartile range lot')
plt.plot(range(2010,2019), s_interq)
```
